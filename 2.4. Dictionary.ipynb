# Dictionary
## to create dict

```python
empty_dict = {}     # empty dictionary by operator '{}'
print(empty_dict)

empty_dict = dict()     # empty dictionary by function 'dict()'
print(empty_dict)
```

```python
tupls_list = [("a", 1), ("b", 2), ("c", 3)]
dict_t = dict(tupls_list) # create dictonary from list of tuples
print(dict_t)

lists_list = [['a', 4], ['b', 5], ['c', 6]]
dict_l = dict(lists_list)   # create dictonary from list of lists
print(dict_l)
```

```python
            ### !!!!!!!!!!!!! ###
keys = ['a', 'b']   # list of keys
values = [1, 2]     # list of values
dict1 = dict(zip(keys, values)) # create dictionary from list of tuples (by 'zip()')
print(dict1)
```

```python
def print_kwargs(**kwargs):         # ** - to # create a dictionary
    print('Named arguments:', kwargs)

print_kwargs(a = 1, b = 2, c = 3)       # create a dictionary!
```

## Chenge dictionary

```python
dict = {'a': 1, 'b': 2, 'c': 3}    # created dict
print(f'created dict - {dict}')

dict.update(d=4, e=5)             # update by providing values
print(f'updated dict - {dict}')

dict.update({'a': 10, 'b' : 20})  # update by another dictionary
print(f'updated dict - {dict}')

dict.update([('c', 30), ('d', 40)])     # update by list of typles
print(f'updated dict - {dict}')

dict.update([['a', 'str'], ['i', 't']])     # update by list of lists
print(f'updated dict - {dict}')
```

```python
del dict['a']   # delete key-value
print(dict)
```

```python
dict.clear()    # delete all elements
print(dict)

# or

dict = {'a': 1, 'b': 2, 'c': 3}
print(dict)
dict = {}       # delete all elements
print(dict)
```

```python
### Create a dictionary with city names as keys and their populations as values ###

Dict = {'Moscow' : 78621,
     'Dolgoprudny' : 6283,
     'Piter' : 82851}
print (f'Created dictionary: {Dict}')

Dict['Rostov'] = 7623  # add a new key-value pair to the dictionary.
print (f'Updated dictionary: {Dict}') # display the updated dictionary.

Dict['Rostov'] = 000    # change a value of key
print(f'Changed dictionary: {Dict}')

print(Dict.items())
print(list(Dict.items()))
```

```python
print(f'Keys: {Dict.keys()}')
print(f'list of keys: {list(Dict.keys())}')
```

```python
print(Dict.values())
print(list(Dict.values()))
```

```python
Plot = {} # empty dictionary
# adding key-value pairs to the dictionary
Plot["ч. 1 ст. 51"] = "шляхом крадіжки вчинив"
Plot['ст. 173'] = 'Фабула ст. 173'
Plot['ч. 1 ст. 173-2'] = "Фабула ч. 1  ст. 173-2"
Plot['ст. 48'] = 'Fabula ct 48'

# iterating through the dictionary to print keys
for key in Plot:
    print(f'key is: {key}')

# Putting all keys into a list
keys = list(Plot.keys())
print(f'keys are: {keys}')

# iterating through the dictionary to print values
for value in Plot:
    print(f'value is: {Plot[value]}')

# Putting all values into a list
values = list(Plot.values())
print(f'values are: {values}')
```

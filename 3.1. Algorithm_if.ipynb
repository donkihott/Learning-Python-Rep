## 4 possible variants

1. else + if = elif

```python
x = 1
y = -1
if x > 0 and y > 0:
    print(1)
elif x < 0 and y > 0:
    print(2)
elif x < 0 and y < 0:
    print(3)
elif x > 0 and y < 0:
    print(4)
else:  # when you did not take into account the option
    print("unexpected option")  # may be x >= 0
```

2.

```python
# unexpect = 0
x = 1
y = -1
if y > 0:
    if x > 0:
        print(1)
    else:           # x < 0
        print(2)
else:               # y < 0
    if x < 0:
        print(3)
    else:           # x > 0
        print(4)
```

3.

```python
# unexpect = 0
x = 1
y = -1
if x > 0 and y > 0:
    print(1)
else:
    if x < 0 and y > 0:
        print(2)
    else:
        if x < 0 and y < 0:
            print(3)
        else:
            print(4)
```

4. with verification if you did not take into account the option

```python
x = 0
y = -1
if x > 0 and y > 0:
    print(1)
else:
    if x < 0 and y > 0:
        print(2)
    else:
        if x < 0 and y < 0:
            print(3)
        else:
            if x > 0 and y < 0:
                print(4)
            else:  # when you did not take into account the option
                print("unexpected option")
```

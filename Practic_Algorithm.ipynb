1. Напишіть  програму,  яка  обчислює  значення  функції  (на  вхід 
подається дійсне число):
$$
Y = 
\begin{cases} 
2x^2 + a & \ x≤0; a=3,5bx; b=-0,9 \\
(x + 3a)c & \ x>0; c=1,35x+b^{\sqrt{x}} 
\end{cases}
$$

```python
x = float(input("input float number:"))
b = - 0.9
a = 3.5 * b * x
 
if x <= 0:
    Y = 2 * (x**2) + a
    print('when x <= 0, Y = ', Y)
else:
    c = (1.35 * x) + (b ** (x **0.5))
    Y = (x + 3 * a) * c
    print('when x > 0, Y = ', Y)
```

2.  Обчислити значення функції на відрізку [x0; xk] з кроком dx 
за допомогою циклу з while.

```python
import matplotlib.pyplot as plt

b = -0.9

def function(current_x):
    a = 3.5 * b * current_x
    if current_x <= 0:
        res_y = 2 * (current_x**2) + a
    else:
        c = (1.35 * current_x) + (complex(b) ** (current_x ** 0.5))
        res_y = (current_x + 3 * a) * c
    return res_y

x0 = float(input("Введите начало x0: "))
xk = float(input("Введите конец xk: "))
dx = float(input("Введите шаг dx: "))

# Списки для хранения координат
x_pts = []
y_pts = []

x = x0
while x <= xk + 1e-9:
    y = function(x)
    x_pts.append(x)
    y_pts.append(y)
    x += dx

# --- Блок графики ---
plt.figure(figsize=(10, 5))

# Строим график действительной части (Real part)
plt.plot(x_pts, [val.real for val in y_pts], label="Real Part of Y", color='blue', linewidth=2)

# Строим график мнимой части (Imaginary part), если она важна
plt.plot(x_pts, [val.imag for val in y_pts], label="Imaginary Part of Y", color='red', linestyle='--')

plt.title("График функции Y(x)")
plt.xlabel("x")
plt.ylabel("Y")
plt.grid(True)
plt.legend()
plt.show()
```

3.   Знайти  суму  n  членів  ряду,  заданого  за  варіантом. 
$$\sum_{i=1}^{n} \frac{\sqrt{2i}}{\sqrt{i+5}-\sqrt{i}}$$

```python
i = 1
n = int(input("Inpunt number of n:"))
total_sum = 0
while i <= n:
    total_sum += (2 * i)**0.5 / ((i+5)**0.5 - i**0.5)
    i += 1
print(f"Сумма {n} членов ряда: {total_sum}")
```

```python
import math

def solve_series_sum():
    # 1. Ввод количества членов ряда
    try:
        n = int(input("Введите количество членов ряда (n): "))
        if n <= 0:
            print("Ошибка: n должно быть больше 0.")
            return
    except ValueError:
        print("Ошибка: введите целое число.")
        return

    total_sum = 0.0

    # 2. Цикл накопления суммы
    for i in range(1, n + 1):
        # Вычисляем числитель и знаменатель отдельно для точности
        numerator = math.sqrt(2 * i)
        denominator = math.sqrt(i + 5) - math.sqrt(i)
        
        # Общий член ряда
        ai = numerator / denominator
        total_sum += ai

    # 3. Вывод результата
    print(f"Сумма {n} членов ряда: {total_sum:.6f}")

solve_series_sum()
```

1. Напишіть  програму,  яка  обчислює  значення  функції  (на  вхід 
подається дійсне число):
$$
Y = 
\begin{cases} 
2x^2 + a & \ x≤0; a=3,5bx; b=-0,9 \\
(x + 3a)c & \ x>0; c=1,35x+b^{\sqrt{x}} 
\end{cases}
$$

```python
x = float(input("input float number:"))
b = - 0.9
a = 3.5 * b * x
 
if x <= 0:
    Y = 2 * (x**2) + a
    print('when x <= 0, Y = ', Y)
else:
    c = (1.35 * x) + (b ** (x **0.5))
    Y = (x + 3 * a) * c
    print('when x > 0, Y = ', Y)
```

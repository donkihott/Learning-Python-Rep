Функція переведення чисел з однієї системи числення до іншої (цілі, беззнакові)

```python
### if number is whole and number > 0 ###
primarybase = 2
number = 11001011
newbase = 16
remainders = [] # list to store remainders
if primarybase > newbase:
    ### Function of successive division using the divmod function ###
    while number > 0:
        divisor = newbase # base of the number system
        number, remainder = divmod(number, divisor) # compute remainder and update number to the quotient
        remainders.append(remainder)   # store remainder
        if number == 0:
            print(f"remainders = {remainders}")
    remainders.reverse() # reverse the list to get correct order
    print(f" in base {divisor}: {remainders}")
else:   # if primarybase < newbase but NOT 16 base system!!!!!!!
    total_decimal = 0  # 1. Створюємо змінну для накопичення суми
    number_list = [int(d) for d in str(number)] # Create a list of integers
    number_list.reverse()
    print(number_list)
    while len(number_list) > 0:
        digit = number_list.pop()  # Get the last digit
        power = len(number_list)
        decimal_value = digit * (primarybase ** power)
        total_decimal += decimal_value # 2. Додаємо до підсумку
        if len(number_list) == 0:
            print(f"number {number} in base {newbase} is {total_decimal}")
```

2. Цілого  числа  без  знаку в  2-х  байтовій  комірці.  Записати відповідь у 16-ій формі.

```python
primarybase = 10
number = 12520
newbase = 2
remainders = [] # list to store remainders
if primarybase > newbase:
    ### Function of successive division using the divmod function ###
    while number > 0:
        divisor = newbase # base of the number system
        number, remainder = divmod(number, divisor) # compute remainder and update number to the quotient
        remainders.append(remainder)   # store remainder
        if number == 0:
            print(f"remainders = {remainders}")
    remainders.reverse() # reverse the list to get correct order
    print(f" in base {divisor}:  {remainders}")
    if len(remainders) < 8:
        remainders = [0]*(8 - len(remainders)) + remainders
        print(f"8-bit representation: {remainders}")
    else:
        remainders = [0]*(16 - len(remainders)) + remainders # add leading zeros to make it 16 bits
        print(f"16-bit representation: {remainders}")
        ### Create list of tetrads (4 bits) from the list of remainders ###
        tetrads = [remainders[i:i+4] for i in range(0, len(remainders), 4)]
        print(f"list tetrads: {tetrads}")
        ### Create dictionary 2 system to 16 system ###
        hex_dict = {
            '0000': '0',
            '0001': '1',
            '0010': '2',
            '0011': '3',
            '0100': '4',
            '0101': '5',
            '0110': '6',
            '0111': '7',
            '1000': '8',
            '1001': '9',
            '1010': 'A',
            '1011': 'B',
            '1100': 'C',
            '1101': 'D',
            '1110': 'E',
            '1111': 'F'
        }
        hex_string = ''
        for tetrad in tetrads:
            tetrad_str = ''.join(str(bit) for bit in tetrad)
            hex_digit = hex_dict[tetrad_str]
            hex_string += hex_digit
        print(f"Hexadecimal representation: {hex_string}")
else:
    print("primarybase is less than newbase")   # if primarybase < newbase
```

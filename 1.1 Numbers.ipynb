```python
primarybase = 10
number = 15
divisor = 2

print(f"Послідовне ділення числа {number} на {divisor}:")

# Цикл триває, поки частка більша за 0
while number > 0:
    # Використовуємо divmod для отримання цілого та залишку за один крок
    number, remainder = divmod(number, divisor)
    
    print(f"Нова частка: {number}, залишок: {remainder}")

print("Ділення завершено, частка дорівнює 0.")
```

1.3. Послідовність залишків від ділення, починаючи з останнього

```python
primarybase = 10
n = 15
b = 2   # base of the number system
remainders = []    # list to store remainders
#  if base of the number system of n is > b
while n > 0:
    n % b   # compute remainder
    remainders.append(n % b)    # store remainder
    print(f"remainder = {n % b}")
    n = n // b  # update n to the quotient
remainders.reverse()   # reverse the list to get correct order
remainders
```

```python
primarybase = 10
number = 203
newbase = 2
remainders = [] # list to store remainders
#  if primarybase of the number system of n is > b
# and number is whole positive, continue
# successive division
# whole, remainder = divmod(number, divisor)

while number > 0:
    divisor = newbase # base of the number system
    number, remainder = divmod(number, divisor) # compute remainder and update number to the quotient
    remainders.append(remainder)   # store remainder
    print(f"remainders = {remainders}")
remainders.reverse() # reverse the list to get correct order
print(f"number in base {divisor}: {remainders}")
```

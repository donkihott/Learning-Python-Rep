```python
## formula choose all numbers from "start" (9) to (0) - "stop"')
x = 9  # start (9)
while x >= 0:  # 'stop' (0)
    print(x)
    x -= 1  # 'step', the came: x = x - 1
```

### iterations with a break in time

```python
import time     # for iterations with a break in time
x = 9  # start (9)
while x >= 0:  # 'stop' (0)
    time.sleep(1)   # iterations with a break in time (1 second)
    print(x)
    x -= 1  # 'step', the came: x = x - 1
```

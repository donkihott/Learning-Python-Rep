```python
## formula choose all numbers from "start" (9) to (0) - "stop"')
x = 9  # start (9)
while x >= 0:  # 'stop' (0)
    print(x)
    x -= 1  # 'step', the came: x = x - 1
```

### iterations with a break in time

```python
import time     # for iterations with a break in time
x = 9  # start (9)
while x >= 0:  # 'stop' (0)
    time.sleep(1)   # iterations with a break in time (1 second)
    print(x)
    x -= 1  # 'step', the came: x = x - 1
```

$$\sum_{i=1}^{n} i^3 = 1^3 + ... + n^3$$

```python
n = int(input('Input n:'))
sum = 0
i = 1
while i <= n:
    sum += i **3    # successively sums the cubes from 'i'
    print('i:', i, ', i**3: ', i**3)
    i += 1      # step for next iteration of i
print('total sum = ', sum)
```

### Endless Cycles

### break!

```python
        ### break ###
i = 1           # 'start' of cycle
while True:
    if i > 5:   # 'stop' (5 in) - cycle limit / otherwise it will be infinite
        break   # interrupt the cycle
    print(i)
    i += 1      # 'step' of iterations
```

### continue!

```python
i = 0
while i <= 10:      # stop = stop + step
    i += 1          # step
    if 3 <= i <= 5:     # whithout 3, 4, 5
        print('...')
        continue        # continue after exclusions
    print(i)
```

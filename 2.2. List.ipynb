## Сopying of list

### $$"="$$

```python
        ### '=' ###
    ### Equation of lists - modifying one list affects the other ###
A = [1, 2, 3, 4, 5]
B = A
A[0] = 1000 # modifying list A (first element)
print(f'List A after modification A: {A}')
print(f'List B after modification A: {B}') # List B also reflects the change in A

B[0] = 2 # modifying list B (first element)
print(f'List A after modification B: {A}') # List A also reflects the change in B
print(f'List B after modification B: {B}')
print(f'A is B? - {A is B}')  # --- IGNORE ---
```

### $$"copy()"$$

```python
        ### '.copy()' ###
    ### Creating a copy of a list - modifying one list does not affect the other ###
A = [1, 2, 3, 4, 5]
B = A
C = A.copy()
print(f'list C: {C}')
print(f'list A: {A}')
print(f'A is C? - {A is C}')
print(f'A == C? - {A == C}')  # --- IGNORE ---
```

```python
print(f'A[1] befor change: {A[1]}')
A[1] = 300 # change second element of list A
print(f'After modification A[1], A: {A}')
print(f'After modification A[1], C: {C} - NOT changed')
print(f'After modification A[1], B: {B}')
B
```

### Nested lists

```python
A = [[1, 2, 3], [4, 5, 6], [7, 8, 9]] # a list of lists
print(A)    # all list
print(A[0])     # first element (list) in list
print(A[0][1])  # sekond element of first list in list

for i in A:     # each element (list)
    print(i)
```

```python
for i in A:
    print()
    for j in i:
        print(j, end = " ")
```

```python
### output the elements of the list of lists ###
for i in range(len(A)): # for each sublist
    for k in range(len(A[i])):  # for each element in the sublist
        print((A[i])[k], end=' ')   # print element with a space
    print()
```

### Nested cycles

```python
i = 1           # start
while i <= 10:      # repetition of table rows
    j = 1
    while j <= 10:  # columns
        print(i * j, end="\t")
        j += 1     
    print()     # performs the transition to new line
    i += 1
```

```python
outer = [1, 2, 3]   # Зовнішній цикл  
inner = [4, 5, 6]   # Вкладений (внутрішній) цикл  
for i in outer:      
    for j in inner:          
        print ('i=', i, 'j=', j)
```

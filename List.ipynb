```python
'''Creating a list and checking its type'''
A = [2, 4, 6, 8, 10]
type(A)  # --- IGNORE ---
```

```python
'''unpacking the list'''
for x in A:
    print(x)
```

```python
''' Equation of lists - modifying one list affects the other '''
B = A
A[0] = 1000 # modifying list A (first element)
print(f'List A after modification A: {A}')
print(f'List B after modification A: {B}') # List B also reflects the change in A

B[0] = 2 # modifying list B (first element)
print(f'List A after modification B: {A}') # List A also reflects the change in B
print(f'List B after modification B: {B}')
print(f'A is B? - {A is B}')  # --- IGNORE ---
```

```python
''' Creating a copy of a list - modifying one list does not affect the other '''
A = [2, 4, 6, 8, 10]
C = A.copy()
print(f'list C: {C}')
print(f'list A: {A}')
print(f'A is C? - {A is C}')
print(f'A == C? - {A == C}')  # --- IGNORE ---
```

```python
print(f'A[1]: {A[1]}')
A[1] = 300
print(f'After modification A[1], A: {A}')
print(f'After modification A[1], C: {C} - NOT changed')
print(f'After modification A[1], B: {B}')
```

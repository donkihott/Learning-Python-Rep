## to Create tuple

```python
empty_tuple = ()    # to crate empty tuple by operator '()'
print(empty_tuple)
```

```python
one_name = 'Alex',      # to create tuple with 1 element with ',' in the end!!!
print(one_name)
names = 'Alex', 'Helen', 'Olga' # to create tuple with few elements without ',' in the end!!!
print(names)
names = ('Alex', 'Helen', 'Olga')
print(names)
```

```python
name_list = ['Alex', 'Helen', 'Olga']
print(name_list)
list_to_tuple = tuple(name_list)
print(list_to_tuple)
```

```python
T = 1, 2, 3, 4
print(type(T)) # <class 'tuple'>
print(T)
```

```python
a, b, c, d = T  # create variables for elements
print(c) # 3
```

```python
a = T[0]    # output variable for first element
print(a) # 1
```

```python
a, b, * rest = T    # create variable 'rest' for elements, another than in 'a', 'b'
print('a is ', a) # 1
print('b is ', b) # 2
print('rest are: ', rest) # [3, 4] - list insert a tuple
print(type(rest))
```

```python
def print_args(*args):      # function accepts any number of parameters
    print('taple of positoznal arguments:', args)

print_args(1, 2, 3, 'Hi!')
```

```python
print(T) # (1, 2, 3, 4) - tuple
print(*T) # 1 2 3 4 - elements
print(*T, sep= ':') # 1:2:3:4 - elements, that separated by ':'
print(*T, sep= ':', end='!\n') # 1:2:3:4!  # \n - new line
```

### variables

```python
one = 1
two = 2

two, one = 1, 2     # chenge variables
print('one:', one, 'two:', two) # one: 2 two: 1

one, two = two, one     # chenge variables
print('one:', one, 'two:', two)     # one: 1 two: 2
```

## to Create tuple

```python
T = 1, 2, 3, 4
print(type(T)) # <class 'tuple'>
print(T)
```

```python
a, b, c, d = T  # create variables for elements
print(c) # 3
```

```python
a = T[0]    # create variable for first element
print(a) # 1
```

```python
a, b, * rest = T    # create variable 'rest' for elements, another than in 'a', 'b'
print('a is ', a) # 1
print('b is ', b) # 2
print('rest are: ', rest) # [3, 4] - list insert a tuple
print(type(rest))
```

```python
print(T) # (1, 2, 3, 4) - tuple
print(*T) # 1 2 3 4 - elements
print(*T, sep= ':') # 1:2:3:4 - elements, that separated by ':'
print(*T, sep= ':', end='!\n') # 1:2:3:4!  # \n - new line
```

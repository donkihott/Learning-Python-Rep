````python
# % pip install pyflowchart
from pyflowchart import Flowchart
from IPython.display import display, Markdown

# 1. Ваш код
code = """
x = 1
y = -1

if x > 0 and y > 0:
    print(1)
elif x < 0 and y > 0:
    print(2)
elif x < 0 and y < 0:
    print(3)
elif x > 0 and y < 0:
    print(4)
else:
    print("Никогда")
"""

# 2. Генерация схемы
fc = Flowchart.from_code(code)

# 3. Преобразование в Mermaid диаграмму
flowchart_text = fc.flowchart()

# Преобразуем формат в валидный Mermaid синтаксис
def convert_to_mermaid(flowchart_text):
    lines = flowchart_text.strip().split('\n')
    mermaid_lines = ["graph TD"]
    
    for line in lines:
        if '=>' in line:
            # Преобразуем определения узлов
            parts = line.split('=>')
            node_id = parts[0].strip()
            node_def = parts[1].strip()
            
            # Парсим определение
            if 'operation:' in node_def:
                label = node_def.replace('operation:', '').replace("'", "").strip()
                mermaid_lines.append(f"    {node_id}['{label}']")
            elif 'condition:' in node_def:
                label = node_def.replace('condition:', '').replace("'", "").strip()
                mermaid_lines.append(f"    {node_id}{{{{{label}}}}}")
            elif 'subroutine:' in node_def:
                label = node_def.replace('subroutine:', '').replace("'", "").strip()
                mermaid_lines.append(f"    {node_id}['[{label}]']")
        elif '->' in line:
            # Преобразуем связи
            line = line.strip()
            if '(yes)->' in line:
                parts = line.split('(yes)->')
                mermaid_lines.append(f"    {parts[0]} -->|yes| {parts[1]}")
            elif '(no)->' in line:
                parts = line.split('(no)->')
                mermaid_lines.append(f"    {parts[0]} -->|no| {parts[1]}")
            else:
                mermaid_lines.append(f"    {line}")
    
    return '\n'.join(mermaid_lines)

mermaid_diagram = convert_to_mermaid(flowchart_text)

display(Markdown(f"""
```mermaid
{mermaid_diagram}
"""))
````

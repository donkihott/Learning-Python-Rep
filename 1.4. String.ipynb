```python
welcome_string = 'Hello World'
print(f"welcome_string('{welcome_string}') -", type(welcome_string)) # welcome_string('Hello World') <class 'str'>
```

```python
### """ Multiline Strings """ ###
multiple_string = """Hello World
I`m a multiline string
that spans multiple lines"""
print('multiple_string("""_""") -', type(multiple_string)) # multiple_string("""_""") - <class 'str'>        
print(multiple_string)
```

```python
### \n for New Line ###
multiple_string = "Hello\nWorld\nI`m a multiline string\nthat spans multiple lines"
print('multiple_string("...") -', type(multiple_string)) # multiple_string("_\n_") - <class 'str'>
print(multiple_string)
```

```python
### \n + \t for New Line and Tab ###
multiple_string = "Hello\n\tWorld\n\tI`m a multiline string\n\tthat spans multiple lines"
print('multiple_string("...") -', type(multiple_string)) # multiple_string("_\n_") - <class 'str'>
print(multiple_string)
```

```python
### \t for Tab Space ###
multiple_string = "Hello\tWorld\tI`m NOT a multiline string\tthat spans multiple lines"
print(multiple_string)
```

```python
multiple_string = "Hello\\World\\I`m NOT a multiline string\\that spans multiple lines"
print(multiple_string)
```

```python
s = r"Hello\I'm a raw string\That doesn't process escape sequences"
print("s(r'...') -", str(type(s))) # s(r'_\_') - <class 'str'>
print(s) # Hello\I'm a raw string\That doesn't process escape sequences
```

```python
s = f'World'    # 'f' - format string
print("s(f'_') -", str(type(s))) # s(f'_') - <class 'str'>
print(s) # World
```

```python
s = "ABC"
x, y, z = s     # unpacking (creates 3 variables for elements in string)
print('x = ', x) # x =  A
print('y = ', y) # y =  B
print('z = ', z) # z =  C
```

```python
s = "ABCDEF..."
x, y, z, *rest = s   # unpacking with rest (creates list for remaining elements)
                    #
print('x =', x) # x = A
print('y =', y) # y = B
print('z =', z) # z = C
print('rest =', rest) # rest =  ['D', 'E', 'F', '.', '.', '.']
print(type(rest)) # <class 'list'>
```

```python
string = 'abcdefghijklmnopqrstuvwxyz'
print(string)
print(string[0])    # a
print(string[1])    # b
print(string[-1])   # z
print(string[0:3])  # abc
```

```python
string = 'abcdefghijklmnopqrstuvwxyz'
print(slice := string[0:5])  # abcde
slice = string[20:]         # uvwxyz (from index 20 to end)
print(slice)
slice = string[:3]          # abc (from start to index 3, not including 3)
print(slice)
slice = string[12:20]    # mnopqrstuv (from index of 'm' to index of 'u', not including 'u')
print(slice)
slice = string[-1:0:-1]  # zyxwvutsrqponmlkjihgfedcba (reversed string)
print(slice)
print(string[::7])     # agu (every 7th character)
print(len(string))    # 26
```

```python
alfabet = 'abcdefghijklmnopqrstuvwxyz'
print(alfabet.index('e'))      # 4
print(alfabet.count('e'))      # 1
print(alfabet.split('i'))   # ['abcdefgh', 'jklmnopqrstu', 'vwxyz']
print(alfabet.split(','))   # ['abcdefghijklmnopqrstuvwxyz']
print(alfabet.replace('a', 'A'))  # Abcdefghijklmnopqrstuvwxyz
print(alfabet.upper())        # ABCDEFGHIJKLMNOPQRSTUVWXYZ
```

```python
letters = 'a, b, c, d, e'
print(letters.split()) # ['a', 'b', 'c', 'd', 'e']
```

```python
print(alfabet.join([alfabet, letters]))
print(letters.strip('a')) # delete 'a' from start and end
print(letters.lower())  # All Letters Lowercase
print(letters.capitalize()) # First Letter Capitalized
print(letters.find('c'))
print(alfabet.title())  # All Words Capitalized
print(alfabet.swapcase()) # Invert Case
```

```python
print(alfabet.center(60)) # Centered String in 60 Characters Width
print(alfabet.ljust(60))  # Left Justified String in 60 Characters Width
print(alfabet.rjust(60))  # Right Justified String in 60 Characters Width
```

```python
### whole number formats ###
print('%s' % 42)  # string format
print('%d' % 42)  # decimal format
print('%x' % 42)  # hexadecimal format
print('%o' % 42)  # octal format
```

```python
print('%f' % 3.141592653589793)  # float format
print('%.2f' % 3.141592653589793)  # float format with 2 decimal places
print('%e' % 3.141592653589793)  # exponential format
print('%E' % 3.141592653589793)  # exponential format with capital E
print('%g' % 3.141592653589793)  # general format
```

```python
n = 42
f = 3.141592653589793
s = 'Hello World'
print('{} {} {}'.format(n, f, s))
print('{2} {0} {1}'.format(f, s, n))
print('{n} {f} {s}'.format(n=n, f=f, s=s))
print(f'{n} {f} {s}')
```

```python
print('{0:10d} {1:10f} {2:10s}'.format(n, f, s)) # fixed width formatting as 10 characters
print(f'{n:10d} {f:10f} {s:10s}') # fixed width formatting as 10 characters using f-string
print(f'{n:<10d} {f:<10f} {s:<10s}') # left aligned within 10 characters
print(f'{n:^10d} {f:^10f} {s:^10s}') # center aligned within 10 characters
print(f'{n:>10d} {f:>10f} {s:>10s}') # right aligned within 10 characters
```

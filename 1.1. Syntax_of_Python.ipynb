```python
# enter any value, number or text
input("Введіть значення змінної:") # message to user to enter variable
# '10' - string type
```

```python
# enter just a whole number (integer), no decimal points, may be positive or negative, no text.
int(input('Введіть ціле число:'))
# 10 - integer type
```

```python
float(input('Введіть число з плаваючою крапкою:'))  # enter a number with decimal point
# 10.5 - float type
```

```python
print(1, 2, 3, 4, 5) # print multiple values separated by space
print(1, 2, 3, 4, 5, sep=',') # print multiple values separated by ','
print(1, 2, 3, 4, 5, sep='; ', end='!') # print multiple values separated by '; ' and end with '!'
```

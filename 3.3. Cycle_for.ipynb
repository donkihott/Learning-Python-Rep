### iterations with a break in time (1 second)

```python
import time     # for iterations with a break in time
for x in 9, 8, 7, 6, 5, 4, 3, 2, 1:  # a lot of nambers
    time.sleep(1)   # iterations with a break in time (1 second)
    print(x)
    print('x**2:', x**2)
```

### "while" vs "for"

#### "while"

```python
numbers = [1, 2, 3, 4, 5]
i = 0   # start

while i <= len(numbers) - 1:    # stop
    print(numbers[i])
    i += 1      # step
```

#### "for"

```python
numbers = [1, 2, 3, 4, 5]
for element in numbers:
    print(element)
```

#### for + break + else

```python
numbers = []
for number in numbers:
    print ('this list have same element -', number)
    break
else:
    # sleeping 'break' means that there are Not element
    print("There are Not elements in list, arn't they?")
```

#### zip()

```python
days = ['Monday', 'Tuesday', 'Wednesday']
fruits = ['banana', 'orange']
drinks = ['coffee', 'tea', 'beer']
for day, fruit, drink in zip(days, fruits, drinks):
    print(day, ': drink', drink, 'eat', fruit)
```

#### range(start, stop, step)
- range(stop, step) - if without start - it begins from '0';
- step == 1 (by default), it may be changed

```python
range(0, 5, 1)      # alone 'range()' - without resalt
```

```python
for i in range(0, 5, 1):    # 0 1 2 3 4  Start including, Stop - NOT including
    print(i)
```

```python
for i in range(0, 5):   # 0 1 2 3 4 ('step' default == 1)
    print(i)
```

```python
for i in range(5):   # 0 1 2 3 4 
                    # 'start' default == 0  - including
                    # 'step' default == 1  - NOT including
    print(i)
```

```python
for i in range(4, -1, -1):
    print(i)
```

```python
# range(start, stop, step)
for i in range(20, 10, -2):
    print(i)
```

### iterations with a break in time (1 second)

```python
import time     # for iterations with a break in time
for x in 9, 8, 7, 6, 5, 4, 3, 2, 1:  # a lot of nambers
    time.sleep(1)   # iterations with a break in time (1 second)
    print(x)
    print('x**2:', x**2)
```

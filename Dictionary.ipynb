# Dictionary

```python
'''Create a dictionary with city names as keys and their populations as values'''

Dict = {'Moscow' : 78621,
     'Dolgoprudny' : 6283,
     'Piter' : 82851}
print (f'Created dictionary: {Dict}')

Dict['Rostov'] = 7623  # add a new key-value pair to the dictionary.
print (f'Updated dictionary: {Dict}') # display the updated dictionary.
```

```python
for key in Dict:
    print(key, Dict[key])
```

```python
for key in Dict:
    print(key, sep= ',', end=',\n') # \n is new line
```

```python
Plot = {} # empty dictionary
# adding key-value pairs to the dictionary
Plot["ч. 1 ст. 51"] = "шляхом крадіжки вчинив"
Plot['ст. 173'] = 'Фабула ст. 173'
Plot['ч. 1 ст. 173-2'] = "Фабула ч. 1  ст. 173-2"
Plot['ст. 48'] = 'Fabula ct 48'

# iterating through the dictionary to print keys
for key in Plot:
    print(f'key is: {key}')

# Putting all keys into a list
keys = list(Plot.keys())
print(f'keys are: {keys}')

# iterating through the dictionary to print values
for value in Plot:
    print(f'value is: {Plot[value]}')

# Putting all values into a list
values = list(Plot.values())
print(f'values are: {values}')
```

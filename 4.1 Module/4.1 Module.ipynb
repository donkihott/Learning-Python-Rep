#### rounding

```python
import math
print(math.ceil(5.49), math.ceil(5.50), math.ceil(5.51)) # always Up
print(math.floor(5.49), math.floor(5.50), math.floor(5.51)) # always Down
print(math.floor(5.49 + 0.5), math.floor(5.50 + 0.5), math.floor(5.51 + 0.5)) # as at school
```

#### square root and raises the Number to the Degree

```python
print('square root of 100 and 25:', 100**0.5, 25**0.5)
import math
print('square root of 100 and 25:', math.sqrt(100), math.sqrt(25))

print('to the degree', math.pow(10, 2), 10 ** 2, math.pow(3, 3), 3 ** 3)
```

```python
import math   
print(math.fabs(10), math.fabs(-10), math.fabs(-12.5))
```

#### remainder from division

```python
import math   
print(math.fmod(10, 5), 10 % 5, math.fmod(10, 3), 10 % 3)
```

## $$random$$

```python
import random
print(random.random()) 
print(random.random()) 
print(random.random())
```

```python
import random
print(random.randint(0, 10))    # int (start, stop)
print(random.randint(0, 10))    # int (start, stop)
```

```python
import random 
print(random.randrange(10))         # from range
print(random.randrange(0, 10))      # from range
print(random.randrange(0, 10, 2))   # from range
```

```python
import random
print(random.choice('string'))          # from string
print(random.choice(['s', 't', 'r']))   # from list
print(random.choice(('s', 't', 'r')))   # from tuple
```

```python
lst = list(range(1, 11))       # create a list
print(lst)

import random
random.shuffle(lst)     # stirs list
print(lst)
```

```python
lst = list(range(1, 11))       # create a list

import random
print(random.sample(lst, 3))    # 3 elements from list
print(lst)                  # list does not change

print(random.sample('string', 3))   # 3 elements from string

tpl = tuple(range(1, 8))        # create a tuple
print(tpl)

print(random.sample(tpl, 3))  # 3 elements from tuple
```
